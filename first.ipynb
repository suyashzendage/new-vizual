```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset (Replace 'car_dekho.csv' with your file path)
df = pd.read_csv('car_dekho.csv')
```

```python
# Display basic information about the dataset
print(df.info())
```

```python
# Display the first few rows to get a sense of the data
print(df.head())
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Get summary statistics of numerical columns
print(df.describe())
```

```python
# Check column names and data types
print(df.dtypes)
```

```python
# Visualize the distribution of 'Selling_Price'
plt.figure(figsize=(8, 6))
sns.histplot(df['Selling_Price'], kde=True, bins=30, color='skyblue')
plt.title('Distribution of Selling Price')
plt.xlabel('Selling Price (in lakhs)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot: Present_Price vs. Selling_Price
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Present_Price', y='Selling_Price', hue='Fuel_Type', data=df)
plt.title('Present Price vs. Selling Price')
plt.xlabel('Present Price (in lakhs)')
plt.ylabel('Selling Price (in lakhs)')
plt.show()
```

```python
# Countplot for Fuel_Type distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='Fuel_Type', data=df, palette='pastel')
plt.title('Count of Fuel Types')
plt.show()
```

```python
# Boxplot: Selling Price by Transmission Type
plt.figure(figsize=(8, 6))
sns.boxplot(x='Transmission', y='Selling_Price', data=df, palette='Set2')
plt.title('Selling Price by Transmission Type')
plt.show()
```
